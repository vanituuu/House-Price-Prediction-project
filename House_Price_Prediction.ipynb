```python
# Supervised learning.
# LinearRegression
#1-find a pattern in old data
#2-straight line
#y=mx+c
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error  # r2
from sklearn.preprocessing import StandardScaler

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')



df = pd.read_csv("house_data_large.csv")
# print(df)
# check shape
print(df.shape)
# information about the data
print(df.info())
print(df.describe())
# is there any null value occur or not
print(df.isnull().sum())


x = df[["Sqft","Bedrooms","Bathrooms","YearBuilt","LocationCode"]] #in a 2-D form
# print(x)

y = df["Price"]
# print(y)



plt.plot(x,y)
plt.xlabel('Sqft')
plt.ylabel('Price')
plt.title("HOUSE DATA")
plt.show()
# standardizes all features (all feature get equal imporatnce,mean = 0,s.d=1)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
# predict the house using the linear regression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size =0.2,random_state=42)
model = LinearRegression()  #model is a variable
model.fit(x_train,y_train)  #model.fit() is a function
# predict the data.
pred = model.predict([[x_test]])

print(f"Based on your {x_test} you may score around{pred}")
# Accuracy of prediction

r2= r2_score(y_test,pred)

rmse=np.sqrt(mean_squared_error(y_test,pred))
print("rootmean square",rmse)
print("Accuracy for this prediction",r2)

# coffeicent feature importance
coefficients = pd.df({
    "Feature": x.columns,
    "Coefficient": model.coef_
})
print("\nFeature Importance:")
print(coefficients)

# predict using array.
new_data = np.array([[2000,5,2,2014,4]])
pred = model.predict(new_data)
# print(int(pred))

print("Predicted House Price:", round(pred[0], 2))
# predict using list.
```
